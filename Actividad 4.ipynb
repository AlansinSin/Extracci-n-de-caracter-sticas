```python
# Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
from google.colab import files
```

```python
# Leer archivo cargado
MR = pd.read_excel('microretailer_mit_lift_lab.xlsx')
MR
```

```python
# Verificar info del DF
MR.info()
```

```python
# Imprimir primeras 2 filas
MR.head(2)
```

```python
# Obtener análisis univariado de las variables categóricas
freq_tbl(MR)
```

```python
# Obtener análisis univariado de una variable categórica en específico
table1 = freq_tbl(MR['214_customer_relationship_tools'])
table1
```

```python
# Obtener un filtro de valores más relevantes de la variables categórica seleccionada
Filtro = table1[table1['frequency']>1]
Filtro
```

```python
# Ajustar indice al dataframe
Filtro_index = Filtro.set_index('214_customer_relationship_tools')
Filtro_index
```

```python
# Realizar grafico de barras del dataframe filtrado
Filtro_index.plot(kind = "bar")
plt.title('Herramientas de relación con los clientes')
plt.xlabel('Tipo de Herramienta')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Realizar grafico de dispersión del dataframe filtrado
Filtro_index.plot("frequency", "cumulative_perc", kind = "scatter")
```

```python
# Realizar grafico de área del dataframe filtrado
Filtro_index.plot(kind = 'area', figsize = (10,4),alpha = 0.5)
```

```python
# Realizar grafico de pastel del dataframe filtrado
Filtro_index["frequency"].plot(kind = 'pie', figsize = (10,5), shadow = True, autopct = "%0.1f %%")
```

```python
# Realizar grafico hexagonal del dataframe filtrado, alternativo al scatter plot
Filtro_index.plot.hexbin("frequency", "cumulative_perc", figsize = (10,5))
```
